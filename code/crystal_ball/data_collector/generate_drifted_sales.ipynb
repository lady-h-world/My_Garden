```python
import pandas as pd
```

```python
df = pd.read_pickle('structured_data/sales.pkl')

# train, test split for df
train_df = df.loc[df['Year'].astype(str) < '2015']
test_df = df.loc[df['Year'].astype(str) == '2015']

train_df.reset_index(inplace=True, drop=True)
test_df.reset_index(inplace=True, drop=True)

print(train_df.shape, test_df.shape)

train_df.head()
```

```python
drifted_test_df = test_df.copy()

drifted_test_df['Sales'] *= 10  # mockup drifted target
drifted_test_df['Store'] = drifted_test_df['Store'][::-1].values  # mockuo drifted feature
```

```python
drifted_df = pd.concat([train_df, drifted_test_df])
drifted_df = drifted_df.sample(frac=1).reset_index(drop=True)  # shuffle the dataframe
print(drifted_df.shape)
drifted_df.head()
```

```python
print(drifted_df.shape)
drifted_df.to_pickle('structured_data/drifted_sales.pkl')
```
